# B- FONCTIONS MODELISATION / REDIMENSIONNEMENT / FEATURISATION :

```
# Fonction Modélisation :

def model_fn():
    
    model = ResNet50(
    include_top=True,
    weights='imagenet',
    input_tensor=None,
    input_shape=(224,224,3),
    pooling=None,
    classes=1000
)
    model.set_weights(bc_model_weights.value)
  
    return model
```

```
# Fonction de Redimensionnement des Images :

def preprocess(content):
    img = PIL.Image.open(io.BytesIO(content)).resize([224, 224])
    arr = img_to_array(img)
    return preprocess_input(arr)
```

```
# Extraction des Features par le Modèle dans un Vecteur :

def featurize_series(model, content_series):
    input = np.stack(content_series.map(preprocess))
    preds = model.predict(input)
    output = [p.flatten() for p in preds]
  
    return pd.Series(output)
```

```
# Fonction Feature :

@pandas_udf('array<float>', PandasUDFType.SCALAR_ITER)
def featurize_udf(content_series_iter):
    model = model_fn()
    for content_series in content_series_iter:
        yield featurize_series(model, content_series)
```

```
# Réduction du Dimensionnement :

spark.conf.set("spark.sql.execution.arrow.maxRecordsPerBatch", "1024")
```

```
# Featurisation de la Dataframe Spark :

features_df = image_df.select(col("path"), col('Label'), featurize_udf("content").alias("features"))
```

```
# Nombre d'Images :

features_df.count()
```

# III PCA PRINCIPAL COMPONENT ANALYSIS / ACP ANALYSE EN COMPOSANTES PRINCIPALES :

# A- VECTORISATION / STANDARDISATION ET RECHERCHE D'UN NOMBRE OPTIMAL DE COMPOSANTES :

```
# Conversion en Vecteur Dense :

transform_vecteur_dense = udf(lambda r: Vectors.dense(r), VectorUDT())

features_df = features_df.withColumn('features_vectors', transform_vecteur_dense('features'))
```

```
# Standardisation PCA :

scaler_std = StandardScaler(inputCol="features_vectors", outputCol="features_scaled", withStd=True, withMean=True)

model_std = scaler_std.fit(features_df)

features_df = model_std.transform(features_df)
```

```
# Fonction Composantes :

def recherche_nb_composante(dataframe, nb_comp=500):
       
    pca = PCA(k = nb_comp,
              inputCol="features_scaled", 
              outputCol="features_acp")
 
    model_pca = pca.fit(dataframe)
    variance = model_pca.explainedVariance
    
    plt.plot(np.arange(len(variance)) + 1, variance.cumsum(), c="purple", marker='o')
    plt.xlabel("Nombre de Composantes")
    plt.ylabel("% Variance")
    plt.show(block=False)
 
    def nb_comp ():
        for i in range(500):
            a = variance.cumsum()[i]
            if a >= 0.95:
                 print("{} Composantes Principales expliquent au moins 95% de la Variance Totale".format(i))
                 break
        return i
 
    k=nb_comp()
  
    return k
```

```
# Renommage :

dataframe_pca = features_df
```

```
# Nombre de Composantes expliquant 95% de la Variance :

n_components = recherche_nb_composante(dataframe_pca)
```

# Interprétations :

# On constate que 182 Composantes sont à conserver.

# B- REDUCTION DE DIMENSION :

```
# Réduction de Dimension PCA :

pca = PCA(k=n_components, inputCol='features_scaled', outputCol='vectors_pca')

model_pca = pca.fit(dataframe_pca)
```

```
# Transformation des Images sur les k Premières Composantes :

dataset_reduction = model_pca.transform(dataframe_pca)
```

```
# Matrice de Confusion :

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.GnBu):

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.

    for i, j in itertools.product(range(cm.shape[0]), 
                                  range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")


    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```
# Prédictions :

y_true = dtc_predictions.select("Classe_Index")
y_true = y_true.toPandas()

y_pred = dtc_predictions.select("prediction")
y_pred = y_pred.toPandas()

cnf_matrix = confusion_matrix(y_true, y_pred, labels=range(10))
```

```
# Visualisation :

sns.set_style("darkgrid")

plt.figure(figsize=(7,7))
plt.grid(False)

plot_confusion_matrix(cnf_matrix, classes=range(10))
```
